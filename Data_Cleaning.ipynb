```python
import pandas as pd
import numpy as np
```

```python
community = pd.read_csv("ReferenceCCAProfiles20132017.csv")
community.head()
```

```python
community.rename(columns={'GEOG':'Community_Name'}, inplace = True)
```

```python
crime = pd.read_csv("cjp_cpd_community_area_level_crime_data_1998-2009.csv")
crime.head()
```

```python
crime.dropna(inplace = True)
crime.shape
```

```python
crime.rename(columns={'CA_Name':'Community_Name'}, inplace = True)
crime = crime[["CA #","Community_Name","Agg_Ass"]]
crime.head()
```

```python
crime.loc[crime.Community_Name == "O'Hare Area","Community_Name"] = "O'Hare"
crime.loc[crime.Community_Name == "Loop","Community_Name"] = "The Loop"
crime.loc[crime.Community_Name == "Ashbum","Community_Name"] = "Ashburn"
```

```python
data = pd.merge(crime,community,how = "left")
data.head()
data.shape
```

```python
# every community name is matched!
data.isnull().loc[data.isnull().UND19 == True,:]
```

```python
data = data[["Community_Name","Agg_Ass","TOT_POP","EMP",
"2000_WHITE",
"2000_HISP",
"2000_BLACK",
"2000_ASIAN",
"2000_OTHER",
"FOR_BORN",
"COMMperc",
"INSTperc",
"VACperc",
"LING_ISO", 
"TOT_COMM",
"WORK_AT_HOME",
"DROVE_AL",
"CARPOOL",
"TRANSIT",
"WALK_BIKE"]]
```

```python
data[["White","Hisp","Black","Asian","Other"]]= data[["2000_WHITE","2000_HISP","2000_BLACK","2000_ASIAN","2000_OTHER"]]
data.drop(["2000_WHITE","2000_HISP","2000_BLACK","2000_ASIAN","2000_OTHER"],axis=1,inplace = True)
```

```python
data["EMP_rate"]= round(data.EMP/data.TOT_POP,2)
data.head()
```

```python
data.to_csv("Final_data.csv")
```
